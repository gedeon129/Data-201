                                          Goal: Predict High Alcohol Wines

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv('/content/winequality-red.csv', sep= ";")
df.head()
```

```python
# Boxplot: Alcohol by High Alcohol Class
plt.figure(figsize=(8, 6))
sns.boxplot(x='high_alcohol', y='alcohol', data=df)
plt.title("Boxplot: Alcohol Distribution by Class")
plt.xlabel("High Alcohol (1=True)")
plt.ylabel("Alcohol Content (%)")
plt.show()
```

                                       Interpretation

The boxplot compares the alcohol content between:

Class 0: wines with low alcohol (< 10.5%)

Class 1: wines with high alcohol ( 10.5%)


The median alcohol content for high-alcohol wines is around 11.5%, while low-alcohol wines average around 9.6%.

High-alcohol wines show more variability, with several outliers exceeding 13% and even 14.9%.

The clear separation between the medians indicates that alcohol content is a strong variable for classification.

Conclusion: Alcohol content clearly separates the two classes, confirming that our binary target (high_alcohol) is meaningful and well-supported by the data.

```python
# Create a binary target: is alcohol >= 10.5%
df['high_alcohol'] = (df['alcohol'] >= 10.5).astype(int)
```

```python
# Select features
features = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
            'chlorides', 'free sulfur dioxide', 'total sulfur dioxide',
            'density', 'pH', 'sulphates', 'quality']
```

```python
X = df[features]
y = df['high_alcohol']
```

```python
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Train the Decision Tree
dt_model = DecisionTreeClassifier(max_depth=4, random_state=42)
dt_model.fit(X_train, y_train)
```

```python
# Predict and evaluate
y_pred = dt_model.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
# Plot the Decision Tree
plt.figure(figsize=(20, 10))
plot_tree(dt_model, feature_names=features, class_names=['Low Alcohol', 'High Alcohol'],
          filled=True, rounded=True, fontsize=10)
plt.title('Decision Tree for High Alcohol Prediction')
plt.show()
```

```python
# Feature Importance
feature_importance = pd.DataFrame({
    'Feature': features,
    'Importance': dt_model.feature_importances_
}).sort_values('Importance', ascending=False)
```

```python
# Plot
plt.figure(figsize=(10, 6))
plt.barh(feature_importance['Feature'], feature_importance['Importance'])
plt.xlabel('Importance')
plt.title('Feature Importance for High Alcohol Prediction')
plt.tight_layout()
plt.show()
```

the feature importance chart reveals which variables the decision tree relied on most when predicting whether a wine has high alcohol content. The standout feature is density, which plays a dominant role in the model. This makes sense because higher alcohol wines tend to be less dense due to the physical properties of ethanol compared to water and sugar.

Quality also contributes significantly to the prediction. This indicates that wines rated higher in quality often contain more alcohol, which aligns with consumer and expert expectations. Citric acid and residual sugar show moderate importance; wines with more citric acid and less sugar are slightly more likely to have higher alcohol levels.

On the other hand, features like chlorides, volatile acidity, and sulphates have very little impact on the models ability to classify alcohol level. While they may influence other aspects of wine quality or flavor, they are not strongly tied to alcohol concentration in this dataset.

In summary, the chart confirms that density is the most reliable indicator of alcohol content, with quality and citric acid offering additional, but smaller, contributions. The remaining features have a much lower influence on classification decisions.
